## Separating my features and target for preparing an model

```python
x=data.drop(['Drug'],axis=1)
x
```

```python
y=data['Drug']
y
```

## Dividing my dataset  into Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=102)
```

## Preparing an Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
obj3=DecisionTreeClassifier()
```

```python
obj3.fit(x_train,y_train)
```

```python
y_train
```

```python
train_predict=obj3.predict(x_train)
train_predict
```

## Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_train,train_predict)
```

```python
print(classification_report(y_train,train_predict))
```

```python
y_test
```

```python
test_predict=obj3.predict(x_test)
test_predict
```

```python
confusion_matrix(y_test,test_predict)
```

```python
print(classification_report(y_test,test_predict))
```

## Random forest and votting classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest=RandomForestClassifier(n_estimators=10)
```

```python
forest.fit(x_train,y_train)
```

```python
trees=forest.estimators_
trees
```

```python
list(enumerate(trees))
```

```python
plt.figure(figsize=(80,60))
for i,j in list(enumerate(trees)):
    plot_tree(j,feature_names=['Age','BP','Cholesterol','Na_to_K'],class_names=['drugA', 'drugB', 'drugC', 'drugX', 'drugY'],filled=True)

plt.show()
```
